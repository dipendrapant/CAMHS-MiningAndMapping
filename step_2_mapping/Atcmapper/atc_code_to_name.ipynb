**Step 1: Make a `JSON` file with Medication Code and Medication Name by extracting all the ATC code and ATC names form the main dataset.**

```python
import os
import sys

parent_dir = os.path.join(
    os.path.dirname(os.path.abspath("")), os.pardir, os.pardir, os.pardir
)
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)
from importall import *

MAIN_DATASET = os.getenv("MAIN_DATASET")
ATCNAME_CODE_NORSK = os.getenv("ATCNAME_CODE_NORSK")
df = dd.read_parquet(MAIN_DATASET, engine="pyarrow")
df_atc = df[["atckode", "atcnavn"]].drop_duplicates().compute()
df_unique = df_atc.groupby("atckode", as_index=False).first()
atc_dict = df_unique.set_index("atckode").to_dict(orient="index")

with open(
    ATCNAME_CODE_NORSK + "atcnavn_code_norsk.json",
    "w",
) as json_file:
    json.dump(atc_dict, json_file, indent=4)
```
